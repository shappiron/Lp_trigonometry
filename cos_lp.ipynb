```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
def cos_l(l):
    C = []
    for a in alpha:
        k = np.abs(np.tan(a))
        x = (1/(1 + k**l))**(1/l)
        y = (1 - x**l)**(1/l)
        sign = np.sign(np.cos(a))
        sign = sign if sign != 0 else 1.0
        cos = sign * x / (x**l + y**l)**(1/l)
        C.append(cos)
    return C

def sin_l(l):
    C = []
    for a in alpha:
        k = np.abs(np.tan(a))
        x = (1/(1 + k**l))**(1/l)
        y = (1 - x**l)**(1/l)
        sign = np.sign(np.sin(a))
        sign = sign if sign != 0 else 1.0
        cos = sign * y / (x**l + y**l)**(1/l)
        C.append(cos)
    return C
```

## Cos

```python
plt.figure(figsize=(9,9))
alpha = np.linspace(0, np.pi/2, 1000)
for l in np.logspace(-0.3, 2, 8):
    C = cos_l(l)
    plt.plot(alpha, C, label='Cos(x) for p={0:.2f}'.format(l))

#plt.plot(alpha, 1-2/np.pi*alpha, color='grey', lw=2, alpha=1)
plt.grid(alpha=0.5)
plt.xlabel('x', fontsize=14)
plt.ylabel('Cos(x)', fontsize=14)
plt.legend(fontsize=14)

plt.show()
```
